```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
import scipy.stats as st
import numpy as np  

# Study data files
state_finance_path = "~/desktop/projects/Group Project 1/states.csv"
naep_path = "~/desktop/projects/Group Project 1/naep.csv"

# Read the data 
state_finance = pd.read_csv(state_finance_path)
naep_results = pd.read_csv(naep_path)

state_finance.head()

```

```python
#create state rev percent series

staterevpercent = state_finance['STATE_REVENUE']/state_finance['TOTAL_REVENUE']

state_finance['staterevpercent'] = staterevpercent

state_finance.head()




```

```python
data2012 = state_finance.loc[(state_finance["YEAR"] == 2012)]
data2012_df = pd.DataFrame(data2012)

data2013 = state_finance.loc[state_finance["YEAR"] == 2013]
data2013_df = pd.DataFrame(data2013)

data2014 = state_finance.loc[state_finance["YEAR"] == 2014]
data2014_df = pd.DataFrame(data2014)

data2015 = state_finance.loc[state_finance["YEAR"] == 2015]
data2015_df = pd.DataFrame(data2015)

data2016 = state_finance.loc[state_finance["YEAR"] == 2016]
data2016_df = pd.DataFrame(data2016)

twoyeardata = pd.merge(data2012_df, data2013_df, on = 'STATE', how = "left")
threeyeardata = pd.merge(twoyeardata, data2014_df, on = 'STATE', how = "left")
fouryeardata = pd.merge(threeyeardata, data2015_df, on = 'STATE', how = "left")
fiveyeardata = pd.merge(fouryeardata, data2016_df, on = 'STATE', how = 'left')

fiveyeardata = pd.DataFrame(fiveyeardata[["STATE","YEAR", "INSTRUCTION_EXPENDITURE","SUPPORT_SERVICES_EXPENDITURE","staterevpercent"]])

fiveyeardata.head()
```

```python
fiveyearmean = fiveyeardata.groupby("STATE").staterevpercent.mean()
fiveyearmean_df = pd.DataFrame(fiveyearmean)
fiveyearmean_df.head()
fiveyearsort = fiveyearmean_df.sort_values(by=['staterevpercent'])
fiveyearsort



#df.sort_values(by=['col1'])





                                                                     
                                
```
